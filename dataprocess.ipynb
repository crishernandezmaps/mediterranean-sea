```python
import pandas as pd
from geopandas import GeoDataFrame
import shapely.geometry
from shapely.geometry import Point
```

```python
df = pd.read_csv('data/missingMigrants.csv')
med = df.loc[(df['Region of Incident'] == 'Mediterranean') & df['Information Reliability'].isin(['Verified'])]
med.head()
```

```python
x = lambda row: float(row.split(",")[0])
y = lambda row: float(row.split(",")[1])
```

```python
med['latitude'] = med['Location'].map(x)
```

```python
med['longitude'] = med['Location'].map(y)
```

```python
med.head()
```

```python
geometry = [Point(xy) for xy in zip((med.longitude).astype(int), (med.latitude).astype(int))]
med = med.drop(['latitude', 'longitude'], axis=1)
crs = {'init': 'epsg:4326'}
gdf = GeoDataFrame(med, crs=crs, geometry=geometry)
```

```python
gdf.to_file("mediterraneum.geojson", driver='GeoJSON')
```
